```python
import requests
from datetime import datetime

# Coordenadas aproximadas para las zonas de Bogotá
ubicaciones = [
    {"nombre": "Zona Norte", "latitud": 4.7488, "longitud": -74.0484},
    {"nombre": "Zona Noroccidental", "latitud": 4.713, "longitud": -74.1067},
    {"nombre": "Zona Nororiental", "latitud": 4.714, "longitud": -74.035},
    {"nombre": "Zona Oriental", "latitud": 4.608, "longitud": -74.0306},
    {"nombre": "Zona Occidental", "latitud": 4.609, "longitud": -74.125},
    {"nombre": "Zona Sur", "latitud": 4.5361, "longitud": -74.0916},
    {"nombre": "Zona Suroccidental", "latitud": 4.533, "longitud": -74.13},
    {"nombre": "Zona Suroriental", "latitud": 4.535, "longitud": -74.05},
    {"nombre": "Zona Centro", "latitud": 4.6097, "longitud": -74.0817},
    {"nombre": "Zona Centro Oriental", "latitud": 4.61, "longitud": -74.06},
    {"nombre": "Zona Centro Occidental", "latitud": 4.61, "longitud": -74.1},
]

# Función para obtener el clima actual
def obtener_clima_actual(latitud, longitud):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={latitud}&longitude={longitud}&current_weather=true&hourly=humidity&timezone=America/Bogota"
    respuesta = requests.get(url)
    
    try:
        datos = respuesta.json()
        # Verificar si el clima actual está disponible
        if 'current_weather' in datos and 'hourly' in datos:
            clima_actual = datos['current_weather']
            humedad = datos['hourly']['humidity'][0]  # Obtener la primera hora de humedad
            
            # Definir "probabilidad de lluvia" como precipitación mayor que cero
            probabilidad_lluvia = "Alta" if clima_actual["precipitation"] > 0 else "Baja"
            
            return {
                "hora": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "temperatura": clima_actual["temperature"],
                "humedad": humedad,
                "precipitacion": clima_actual["precipitation"],
                "probabilidad_lluvia": probabilidad_lluvia
            }
        else:
            print("Datos actuales no disponibles en la respuesta de la API.")
            return None
    except Exception as e:
        print(f"Error procesando datos: {e}")
        return None

# Obtener y mostrar el clima actual para cada zona
for ubicacion in ubicaciones:
    print(f"\nClima actual para {ubicacion['nombre']}:")
    clima_actual = obtener_clima_actual(ubicacion["latitud"], ubicacion["longitud"])
    
    if clima_actual:
        print(f"  Hora: {clima_actual['hora']}")
        print(f"  Temperatura: {clima_actual['temperatura']}°C")
        print(f"  Humedad: {clima_actual['humedad']}%")
        print(f"  Precipitación: {clima_actual['precipitacion']} mm")
        print(f"  Probabilidad de lluvia: {clima_actual['probabilidad_lluvia']}")
    else:
        print("  No se pudo obtener el clima para esta ubicación.")
    print("-" * 50)
```
