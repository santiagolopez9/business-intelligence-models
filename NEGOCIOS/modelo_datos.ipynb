```python
import pandas as pd

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos.csv')

# Separar los datos de las etiquetas (precios de venta)
x_train = data.iloc[:,:3].values
y_train = data.iloc[:,3].values
```

```python
from sklearn.linear_model import LinearRegression

# Entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(x_train, y_train)

# Hacer una predicción para 400 unidades vendidas en los últimos 3 meses
units_to_sell = [[350, 450, 550]]
predicted_price = model.predict(units_to_sell)

print("Para vender 400 unidades en el mes siguiente se debería vender a un precio promedio de: ${:.2f}".format(predicted_price[0]))
```
